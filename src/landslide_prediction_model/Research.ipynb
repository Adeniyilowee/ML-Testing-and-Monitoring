```python
# imputation categorical variables
imputer = SimpleImputer(strategy='most_frequent')
df[['LITHOLOGY', 'SCARPS', 'WOODS']] = imputer.fit_transform(df[['LITHOLOGY', 'SCARPS', 'WOODS']])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
import pandas as pd
from sklearn.svm import SVC
```

```python
data = {"Name": ["Fiona", "Gerald", "Hans", "Isabella", "Jacob"],
        "Age": [20, 34, None, None, 33],
        "Gender": ["f", "m", "m", "f", "m"],
        "Job": ["Writer", "Programmer", "Programmer", "Programmer", "Teacher"]}
data= pd.DataFrame(data)
```

```python
class NameDropper(BaseEstimator, TransformerMixin):
    
    def fit (self, X, y=None):
        
        return self
        
    def transform (self, X):
        
        return X.drop(['Name'], axis=1)
        
class AgeImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        
        return self
        
    def transform(self, X):
        imputer = SimpleImputer(strategy="mean")
        X['Age'] = imputer.fit_transform(X[['Age']])
        X_ = X.copy()
        X_.to_csv('cool.csv')
        return [X, X_]

class FeatureEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        
        return self
        
    def transform(self, X):
        gender_dct = {"m": 0, "f": 1}
        X[0]["Gender"] = [gender_dct[g] for g in X[0]["Gender"]]
        
        encoder = OneHotEncoder()
        matrix = encoder.fit_transform(X[0][['Job']]).toarray ()
        column_names = ["Programmer", "Writer", "Cook", "Teacher"]
        for i in range(len(matrix.T)):
            X[0][column_names[i]] = matrix.T[i]
        X[0].drop(['Job'], axis=1, inplace=True)
        return X[0]#, X[1]

class Model(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.classifier_svm_ = SVC(C = 113, gamma= 0.12, kernel ='rbf', probability=True)
    
    def fit(self, X, y=None):
        
        return self
        
    def transform(self, X):
        self.classifier_svm_.fit(X[['Age', 'Programmer', 'Writer', 'Cook']], X['Gender'])
        
        return self.classifier_svm_
```

```python
pipe = Pipeline([("dropper", NameDropper()), ("imputer", AgeImputer()), ("encoder", FeatureEncoder())])
O, T = pipe.fit_transform(data)
```

```python
pipe = Pipeline([("dropper", NameDropper()), ("imputer", AgeImputer()), ("encoder", FeatureEncoder()),  ("mod", Model())])
mod = pipe.fit_transform(data)
```
